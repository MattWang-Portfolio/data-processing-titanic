```python
import matplotlib.pyplot as plt
import matplotlib.style
import pandas as pd
import numpy as np

matplotlib.style.use('ggplot')
```

```python
df = pd.read_csv('http://bit.ly/kaggletrain')
```

```python
df
```

```python
fig, axes = plt.subplots(2, 2, figsize=(15, 15))
x=[1, 2]
y=df[df.loc[:, "Sex"]=="male"].count()
z=df[df.loc[:, "Sex"]=="female"].count()
a=[y.Sex,z.Sex]
labels = ['male','female']
axes[0,0].bar(x,a,tick_label=labels)

b=df["Age"]
axes[0,1].hist(b,bins=20)

z=[1,2,3]
p1=df[df['Pclass']==1]
p1_live=p1[p1['Survived']==1]
p1_dead=p1.shape[0]-p1_live.shape[0]
p2=df[df['Pclass']==2]
p2_live=p2[p2['Survived']==1]
p2_dead=p2.shape[0]-p2_live.shape[0]
p3=df[df['Pclass']==3]
p3_live=p3[p3['Survived']==1]
p3_dead=p3.shape[0]-p3_live.shape[0]
c1=[p1_dead,p2_dead,p3_dead]
c2=[p1_live.shape[0],p2_live.shape[0],p3_live.shape[0]]
labels2=['Pclass1','Pclass2','Pclass3']
width=0.4
axes[1,0].bar(z ,c1 ,width=width, tick_label=labels2,label='Dead')
z2 = [ num + width for num in z]
axes[1,0].bar(z2 , c2 ,width=width ,label='Live' )
axes[1,0].legend(loc='upper left')

die=df[df. loc[:,"Survived"]== 0]
live=df[df. loc[:,"Survived"]== 1]
axes [1, 1].scatter(die.Age, die.Fare, marker='x',label=' x')
axes [1, 1].scatter(live.Age, live.Fare, marker='s',label=' square')

plt.show()

```

```python
fig, axes = plt.subplots(2, 2, figsize=(15, 15)) #造一個畫布為(2,2)矩陣 15*15的大小
x=[1,2] #設兩區
y=df[df.loc[:, "Sex"]=="male"].count()#設一個y從資料中選取Sex欄位==male的數量
z=df[df.loc[:, "Sex"]=="female"].count()#設一個z從資料中選取Sex欄位==fmale的數量
a=[y.Sex,z.Sex]#長條圖中用到的數據
labels = ['male','female'] #設置長條圖底下的標籤
axes[0,0].bar(x,a,tick_label=labels)#長條圖顯示在(0,0)位置

b=df["Age"] #設b去讀取資料中Age的欄位
axes[0,1].hist(b,bins=20) #總數為20條的直方圖

z=[1,2,3] #設成三個區域
p1=df[df['Pclass']==1] #設p1讀資料裡的Pclass為1
p1_live=p1[p1['Survived']==1] #設定資料內Survived為1
p1_dead=p1.shape[0]-p1_live.shape[0] #設定死亡資料為全部資料減掉Survived資料
p2=df[df['Pclass']==2] #設p2讀資料裡的Pclass為2
p2_live=p2[p2['Survived']==1] #設定資料內Survived為1
p2_dead=p2.shape[0]-p2_live.shape[0] #設定死亡資料為全部資料減掉Survived資料
p3=df[df['Pclass']==3] #設p3讀資料裡的Pclass為3
p3_live=p3[p3['Survived']==1] #設定資料內Survived為1
p3_dead=p3.shape[0]-p3_live.shape[0] #設定死亡資料為全部資料減掉Survived資料
c1=[p1_dead,p2_dead,p3_dead] #這邊設c1為死亡數
c2=[p1_live.shape[0],p2_live.shape[0],p3_live.shape[0]] #這邊設c2為存活數
labels2=['Pclass1','Pclass2','Pclass3'] #設定底下顯示的標籤
width=0.4 #設定寬度為0.4
axes[1,0].bar(z ,c1 ,width=width, tick_label=labels2,label='Dead') #長條圖的語法，裡面也把下面的標籤改成'Dead'
z2 = [ num + width for num in z] #避開c1
axes[1,0].bar(z2 , c2 ,width=width ,label='Live' ) #長條圖語法，裡面把下面標籤改成'Live'
axes[1,0].legend(loc='upper left') #這個語法是把label設定在左上角

die=df[df. loc[:,"Survived"]== 0] #設die讀資料中的Survived==0
live=df[df. loc[:,"Survived"]== 1] #設live讀資料中的Survived==1
axes [1, 1].scatter(die.Age, die.Fare, marker='x',label=' x') #在(1,1)的位置上用分散圖顯示die.Age跟die.Fare，並用x表示
axes [1, 1].scatter(live.Age, live.Fare, marker='s',label=' s')#在(1,1)的位置上用分散圖顯live.Age跟live.Fare，並用s(方形)表示
```
