```python
import matplotlib.pyplot as plt
import matplotlib.style
import pandas as pd
import numpy as np

matplotlib.style.use('ggplot')
```

```python
df = pd.read_csv('http://bit.ly/kaggletrain')
```

```python
df
```

```python
fig, axes = plt.subplots(2, 2, figsize=(15, 15))
x=[1, 2]
y=df[df.loc[:, "Sex"]=="male"].count()
z=df[df.loc[:, "Sex"]=="female"].count()
a=[y.Sex,z.Sex]
labels = ['male','female']
axes[0,0].bar(x,a,tick_label=labels)

b=df["Age"]
axes[0,1].hist(b,bins=20)

z=[1,2,3]
p1=df[df['Pclass']==1]
p1_live=p1[p1['Survived']==1]
p1_dead=p1.shape[0]-p1_live.shape[0]
p2=df[df['Pclass']==2]
p2_live=p2[p2['Survived']==1]
p2_dead=p2.shape[0]-p2_live.shape[0]
p3=df[df['Pclass']==3]
p3_live=p3[p3['Survived']==1]
p3_dead=p3.shape[0]-p3_live.shape[0]
c1=[p1_dead,p2_dead,p3_dead]
c2=[p1_live.shape[0],p2_live.shape[0],p3_live.shape[0]]
labels2=['Pclass1','Pclass2','Pclass3']
width=0.4
axes[1,0].bar(z ,c1 ,width=width, tick_label=labels2,label='Dead')
z2 = [ num + width for num in z]
axes[1,0].bar(z2 , c2 ,width=width ,label='Live' )
axes[1,0].legend(loc='upper left')

die=df[df. loc[:,"Survived"]== 0]
live=df[df. loc[:,"Survived"]== 1]
axes [1, 1].scatter(die.Age, die.Fare, marker='x',label=' x')
axes [1, 1].scatter(live.Age, live.Fare, marker='s',label=' square')

plt.show()

```
